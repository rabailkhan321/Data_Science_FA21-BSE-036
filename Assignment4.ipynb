```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

import pandas as pd
```

```python
s1 = "data science is one of the most important courses in computer science"
s2 = "this is one of the best data science courses"
s3 = "the data scientists perform data analysis"
```

```python
data = [s1,s2,s3]
data
```

```python
#Bag of words
count_vectorizer = CountVectorizer()
c_vector_matrix = count_vectorizer.fit_transform(data)
c_vector_matrix
```

```python
c_tokens = count_vectorizer.get_feature_names_out()
c_tokens
```

```python
dataframebow = pd.DataFrame(data = c_vector_matrix.toarray(), columns = c_tokens)
dataframebow
```

```python
#Term frequency
tf_matrix_exp1_normalized =c_vector_matrix.toarray() / c_vector_matrix .sum(axis=1).reshape(-1, 1)
df_tf_normalized = pd.DataFrame(tf_matrix_exp1_normalized, columns=c_tokens)
print(df_tf_normalized)
```

```python
#idf
tf_vect = TfidfVectorizer(use_idf=True)
idf_matrix = tf_vect.fit_transform(data)
terms = tf_vect.get_feature_names_out()

idf_values = np.log(len(data) / ( 1+np.sum(idf_matrix > 0, axis=0)))

idf_values = np.array(idf_values).reshape(-1)
df_idf = pd.DataFrame(idf_values, index=terms, columns=["IDF"])
print(df_idf)
```

```python
#TF-IDF
tfidf_vect = TfidfVectorizer()
t_tfidf_matrix= tfidf_vect.fit_transform(data)
t_tfidf_matrix
```

```python
t_tokens = tfidf_vect.get_feature_names_out()
t_tokens
```

```python
df_t_exp1 = pd.DataFrame(data = t_tfidf_matrix.toarray(), columns = t_tokens)
df_t_exp1
```

```python
#cosine:
#bow
t_cosine_similarity_matrix_bow = cosine_similarity(c_vector_matrix)

df_cosine_bow = pd.DataFrame(t_cosine_similarity_matrix_bow, columns=["s1", "s2", "s3"], index=["s1", "s2", "s3"])
df_cosine_bow
```

```python
#cosine:
#tf
tf_matrix_exp1_normalized_array = np.asarray(tf_matrix_exp1_normalized)

t_cosine_similarity_matrix_tf = cosine_similarity(tf_matrix_exp1_normalized_array)

print(t_cosine_similarity_matrix_tf)
```

```python
#cosine:
#idf
idf_matrix_array = idf_matrix.toarray()  # Convert the sparse matrix to a NumPy array
t_cosine_similarity_matrix_idf = cosine_similarity(idf_matrix_array)
print(t_cosine_similarity_matrix_idf)
```

```python
#cosine:
#tfidf
t_cosine_similarity_matrix_tfidf = cosine_similarity(t_tfidf_matrix)
t_cosine_similarity_matrix_tfidf
df_t_similarity = pd.DataFrame(data = t_cosine_similarity_matrix_tfidf)
df_t_similarity
```
